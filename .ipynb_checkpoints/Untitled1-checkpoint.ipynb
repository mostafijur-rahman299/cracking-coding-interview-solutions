```python
class Node:
    def __init__(self, val, next_node=None):
        self.val = val
        self.next = next_node

        
class LinkedList:
    def __init__(self, head=None):
        self.head = head
        
    def print_node(self):
        current = self.head
        while current:
            print(current.val)
            current = current.next
        
    def append_item(self, data):
        new_node = Node(data)
        
        if not self.head:
            self.head = new_node
            return
        
        current = self.head
        while current.next:
            current = current.next
            
        current.next = new_node
        
    def prepend_item(self, data):
        new_node = Node(data)
        
        if not self.head:
            self.head = new_node
            return
        
        temp = self.head
        self.head = new_node
        self.head.next = temp
        
    def append_middle_position(self, data):
        new_node = Node(data)
        
        if not self.head:
            self.head = new_node
            return
        
        current = self.head
        runner = self.head
        
        while current and runner.next.next:
            runner = runner.next.next
            current = current.next
            
            
        cur_next = current.next
        current.next = new_node
        current.next.next = cur_next
        
    def remove_head_node(self):
        if not self.head:
            return
        
        cur_next = self.head.next
        self.head = cur_next
        
    def remove_last_node(self):
        if not self.head:
            return
        
        current = self.head
        prev = self.head
        
        while current.next:
            prev = current
            current = current.next
            
        prev.next = None
        
    def remove_middle_node(self):
        if not self.head:
            return
        
        current = self.head
        runner = self.head
        curr_prev = self.head
        
        while current and runner.next:
            runner = runner.next.next
            curr_prev = current
            current = current.next
            
        curr_prev.next = current.next
        
        
    def remove_item(self, data):
        if not self.head:
            return
        
        current = self.head
        while current.next:
            if current.val == data:
                return
            
            current = current.next
            
        print(current.val)
            
        
        

linked_list = LinkedList()


linked_list.append_item(10)
linked_list.append_item(11)
linked_list.append_item(12)

# linked_list.print_node()

linked_list.prepend_item(9)
linked_list.prepend_item(8)
linked_list.prepend_item(7)

# linked_list.print_node()

linked_list.append_middle_position(11)

linked_list.remove_head_node()

linked_list.remove_last_node()

linked_list.remove_middle_node()

linked_list.remove_item(8)

linked_list.print_node()
```
